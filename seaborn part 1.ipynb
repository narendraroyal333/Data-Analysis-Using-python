```python
import matplotlib.pyplot as plt
import numpy as np
x= np.linspace(0,14,100) # 100 parts between 1 and 14
for i in range(1,5):
    plt.plot(x,np.sin(x+i*0.5)*(7-i))
sns.set_style("darkgrid")    
plt.show()    
    
```
