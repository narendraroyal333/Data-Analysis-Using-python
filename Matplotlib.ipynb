# Matplotlib 
 - Matplotlib is one of the library in python used for 2D plotting in visualization
 - Pyplot
  - Pyplot is a collection of functions.
  - Each functions can create a plotting area in figure,plot lines,plot with labels

## Types of plotting in matplot
- lines
- Bar chart
- Scatter plot
- Histogram
- Box plot
- Pie chart

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Line Plotting
 - plt.plot()

```python
x_axes=[1,2,3,4]
y_axes=[1,4,6,8]
plt.plot(x_axes,y_axes,'r')
plt.show()
```

```python
x_axes=[1,2,3,4]
y_axes=[1,4,6,8]
plt.plot(x_axes,y_axes,'r')
plt.xlabel('X axes')
plt.ylabel('Y axes')
plt.title("Line")
plt.show()
```

```python
x=np.linspace(0,2,10)
x
```

```python
plt.plot(x,x,label='Line')
plt.plot(x,x**2,label='Square')
plt.legend(loc='best')
```

```python
help(plt.legend)
```

```python
plt.xticks([2,4,6,8])
plt.yticks([2,4,6,8])
plt.show()
```

# Bar Chart
 - plt.bar()

```python
plt.bar([1,2,3,4,5],[2,6,4,8,10],width=0.33,color='k')
plt.show()
```

```python
x=np.arange(4)
x
```

```python
plt.bar(x,[89,90,55,60],color='b',width=0.33)
plt.bar(x+0.33,[85,88,96,48],color='y',width=0.33)
plt.show()
```

# Histogram
- Histogram is similar to bar chart
 - used for frequency distribution
- plt.hist() 

```python
a=[0,0,0.2,0.3,2,2.2,2.5,4,4.1,4.4,5.1]
b=5
plt.hist(a,b,rwidth= 0.4)
plt.show()
```

```python
a2=[22,22,23,23,24,25,25,26,26,28.3,28.4,28.6,30,30.1,30.2]
bin2=[21,23,24,25,26,27,28,29,31]
plt.hist(a2,bin2,rwidth=0.4)
plt.show()
```

# Boxplot
- plt.boxplot()

```python
marks=np.random.randint(10,100,101)
marks
```

```python
plt.boxplot(marks)
plt.show()
```

# pie chart 
- plt.pie()

```python
s=['Data analysis','Machine Learning','python']
marks=[80,90,75]
plt.pie(marks,colors=['b','g','y'],labels=s,autopct='%1.2f%%',startangle=90)
plt.show()
```

```python
plt.scatter([1,2,3,4],[2,4,6,8],marker='d')
plt.show()
```
