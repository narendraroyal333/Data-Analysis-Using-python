
- lmplot()
- It creates a regression line between 2 parameters and helps to visualize their linear relationship

```python
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import seaborn as sns
sns.lmplot(x="total_bill",y="tip",data=tips)
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="sex",markers=["o","v"])
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="sex",markers=["o","v"],palette="BuPu_r",scatter_kws={"s":100})
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="sex",markers=["o","v"],palette="BuPu_r")
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,col="sex",row="time",hue="smoker")
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,col="sex",row="time",hue="smoker",height=6,palette="coolwarm")
```

```python
#bin represents how many falls in data
sns.distplot(tips["total_bill"],bins=10)
```
