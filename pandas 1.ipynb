# Data Frames

- varoius Format of files can able to read using data frames
  - files like excel,csv ,jsonetc

```python
import pandas as pd
pd.__version__
```

## create Data frames

```python
d1=pd.DataFrame([1,2,3,4])
d1 # default index from 0
```

```python
d2=pd.DataFrame([["abc",1,2],["dfg",4,5]])
# data frame with 2 columns
d2
```

```python
# data frame with column names
d2=pd.DataFrame([["abc",1,2020],["dfg",4,2020]],columns=("name","rollno","year"))
d2
```

```python
d2=pd.DataFrame([["abc",1,2020],["dfg",4,2020]],columns=("name","rollno","year"),index=["10","11"]) # to give particular index
d2
```

```python
dict1={"Name":["abc","dfg","zxc"],"rollno":[1,2,3],"year":[2015,2016,2020]}
dict1
```
